 1. What is the difference between a function and a method in Python?
 -  Function

- A function in Python is a block of reusable code that performs a specific task.

- Defined using the def keyword (or lambda for anonymous functions).

- Can exist independently (not tied to any object).

- Can be called directly by its name.

Example of a function:  def greet(name):
    return f"Hello, {name}!"

print(greet("Shivani"))

Method

- A method is a function that is associated with an object (or class).

- It is defined inside a class.

- Always takes self (reference to the object) as its first parameter when it’s an instance method.

- Called using dot notation on an object.

 Example of a method:  class Person:
    def greet(self, name):
        return f"Hello, {name}! I am {self}."

p = Person()
print(p.greet("Shivani"))

5. What are iterators in Python and how do they differ from iterables?
-  Iterable

- An iterable is any Python object that can be looped over.

Examples: lists, tuples, strings, sets, dictionaries.

Technically, an iterable is any object that implements the __iter__() method.

Example: nums = [1, 2, 3]   # list is iterable

    print(n)

- An iterator is an object that produces items one at a time using the __next__() method.

It remembers its state so it doesn’t start over each time.

Created from an iterable using iter().

Example: nums = [1, 2, 3]     # iterable
it = iter(nums)      # make iterator

print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3

- Very useful when working with large datasets or infinite sequences.

Generators are defined in two main ways:
 -  Using a function with the

 7. What are the advantages of using generators over regular functions.
 -  Advantages of Generators over Regular Functions
1. Memory Efficiency

- Regular functions (with return) build the entire result in memory.

- Generators yield one value at a time, so they don’t need to store everything.

# Regular function
def squares_list(n):
    return [i*i for i in range(n)]

# Generator function
def squares_gen(n):
    for i in range(n):
        yield i*i


2. Lazy Evaluation

- Generators compute values only when needed.

- This avoids unnecessary work.

gen = (i*i for i in range(5))
print(next(gen))  # Only calculates first value (0)


3. Infinite Sequences

A normal function cannot return infinite values, but a generator can.

def infinite_numbers():
    n = 0
    while True:
        yield n
        n += 1

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
-   map()

 Purpose: Apply a function to each element in an iterable.
 Returns: A new iterable (map object).

 Example:  numbers = [1, 2, 3, 4, 5]
result = map(lambda x: x*2, numbers)
print(list(result))

 Output:  [2, 4, 6, 8, 10]

- filter()

 Purpose: Select elements from an iterable based on a condition (True/False).
 Returns: A filtered iterable (filter object).

 Example:  numbers = [1, 2, 3, 4, 5, 6]
result = filter(lambda x: x % 2 == 0, numbers)  # keep only even numbers
print(list(result))

 Output:  [2, 4, 6]

-  reduce()

 Purpose: Apply a function cumulatively to the elements of an iterable (reduce them to a single value).
 Returns: A single value.
 Note: reduce() is not a built-in function; it’s in the functools module.

 Example:

from functools import reduce

numbers = [1, 2, 3, 4, 5]
result = reduce(lambda x, y: x + y, numbers)  # sum of all numbers
print(result)


Output:

15

```python
#1 . Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:  # check if number is even
            total += num
    return total

# Example usage
nums = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_of_even(nums))  # Output: 12

```

```python
#4 . Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):  # check till square root
        if num % i == 0:
            return False
    return True

# Print all prime numbers from 1 to 200
for n in range(1, 201):
    if is_prime(n):
        print(n, end=" ")

```

```python
#5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            if self.count == 0:
                self.count += 1
                return 0
            elif self.count == 1:
                self.count += 1
                return 1
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
            raise StopIteration


# Example usage:
fib = Fibonacci(10)  # generate 10 terms
for num in fib:
    print(num, end=" ")
```

```python
#6 Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in power_of_two(5):
    print(power)
```

```python
#7  Implement a generator function that reads a file line by line and yields each line as a string.
 # Step 1: Create a sample file
with open("sample.txt", "w") as f:
    f.write("Hello\n")
    f.write("This is Ritesh\n")
    f.write("Learning Python\n")
 # Step 2: Generator function to read file line by line
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()   # strip() removes \n at the end
 # Step 3: Use the generator
for line in read_file_line_by_line("sample.txt"):
    print(line)
```

```python
#9  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# Celsius to Fahrenheit conversion formula: F = (C * 9/5) + 32

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# List of Celsius temperatures
celsius = [0, 20, 37, 100]

# Convert using map
fahrenheit = list(map(celsius_to_fahrenheit, celsius))

print("Celsius:    ", celsius)
print("Fahrenheit: ", fahrenheit)
```

```python
#10 Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, s))

# Example usage
text = "Hello World"
result = remove_vowels(text)
print("Original String:", text)
print("Without Vowels:", result)
```
